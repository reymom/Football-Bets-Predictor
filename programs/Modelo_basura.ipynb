```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Datos/2018-2019.1eraDivision(FT,HTresults;matchstats;match,totalgoals,AHodds).csv')
my_features = ['HomeTeam', 'AwayTeam', 'B365H', 'B365A', 'B365D', 'BWH', 'BWA', 'IWH',
               'IWA', 'PSH', 'PSA', 'WHH', 'WHA', 'VCH', 'VCA' , 'BbAvH', 'BbAvA', 'BbMxH', 'BbMxA', 'FTR']
data = data[my_features]
data.head()
```

```python
def accuracy_profits(data, home_win, away_win):
    accuracy = 0
    profits = 0
    for i in range(len(data)):
        if (data['FTR'][i] == 'H') and (data[home_win][i] < data[away_win][i]):
            accuracy += 1
            profits += (data[home_win][i])

        if (data['FTR'][i] == 'A') and (data[away_win][i] < data[home_win][i]):
            accuracy += 1
            profits += (data[away_win][i])

        if data[away_win][i] == data[home_win][i]:
            print('Misma probabilidad!')
    
    print(' ')
    print('Accuracy = {}'.format(accuracy/len(data)))
    print('Profits = {}'.format(profits))
    print('Win (+ or lost -) = ', profits-len(data))
```

```python
data = pd.read_csv('Datos/2018-2019.1eraDivision(FT,HTresults;matchstats;match,totalgoals,AHodds).csv')
my_features = ['HomeTeam', 'AwayTeam', 'B365H', 'B365A', 'BbAvH', 'BbAvA', 'FTR']
data = data[my_features]
data.head()
```

```python
data.columns
```

```python
def get_data(data):
    X = data.values[:, -3:-1]
    out = data.values[:, -1]
    y_full = np.zeros((X.shape[0], 6))
    for i, y_i in enumerate(out):
        if y_i == 'H':
            y_full[i, 0] = 1.0
        if y_i == 'A':
            y_full[i, 1] = 1.0
        if y_i == 'D':
            y_full[i, 2] = 1.0
        
        y_full[i, 4] = data.values[i, 2] # ODDS OF HOME TEAM in the house you want to bet in
        y_full[i, 5] = data.values[i, 3] # ODDS OF AWAY TEAM blabla
    
    return X, y_full, out
X, y, outcome = get_data(data)
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y, = train_test_split(X,  y) # SPLIT THE DATA IN TRAIN AND TEST DATASET.
```

```python
from keras.layers import BatchNormalization, Dense, Input, Dropout
from keras.models import Model
from keras import backend as K

def get_model(input_dim, output_dim, base=1000, multiplier=0.25, p=0.2):
    inputs = Input(shape=(input_dim,))
    l = BatchNormalization()(inputs)
    l = Dropout(p)(l)
    n = base
    l = Dense(n, activation='relu')(l)
    l = BatchNormalization()(l)
    l = Dropout(p)(l)
    n = int(n * multiplier)
    l = Dense(n, activation='relu')(l)
    l = BatchNormalization()(l)
    l = Dropout(p)(l)
    n = int(n * multiplier)
    l = Dense(n, activation='relu')(l)
    outputs = Dense(output_dim, activation='softmax')(l)
    model = Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer='Nadam', loss=odds_loss)
    return model
```

```python
def odds_loss(y_true, y_pred):
    """
    The function implements the custom loss function
    
    -Inputs
    true : a vector of dimension batch_size, 7.
           A label encoded version of the output and the backp1_a and backp1_b
    pred : a vector of probabilities of dimension batch_size , 5.
    
    -Returns 
    the loss value
    """
    
    win_home = y_true[:, 0:1]
    win_away = y_true[:, 1:2]
    draw = y_true[:, 2:3]
    no_bet = y_true[:, 3:4]
    
    odds_a = y_true[:, 4:5]
    odds_b = y_true[:, 5:6]
    
    gain_loss_vector = K.concatenate([win_home * (odds_a - 1) - (1 - win_home),
                                      win_away * (odds_b - 1) - (1 - win_away),
                                      draw * (1/(1 - 1/odds_a - 1/odds_b) - 1) - (1 - draw),
                                      K.zeros_like(odds_a)], axis=1)

    return -1 * K.mean(K.sum(gain_loss_vector * y_pred, axis=1))

# true = K.variable(np.array([[1, 0, 0, 0, 2.0, 3.0]]), dtype='float32')
# pred = K.variable(np.array([[0.6, 0.1, 0.2, 0.05]]), dtype='float32')

# K.eval(odds_loss(true, pred))
```

```python
from keras.callbacks import EarlyStopping, ModelCheckpoint

model = get_model(2, 4, 1000, 0.9, 0.7)
history = model.fit(train_x, train_y, 
                    validation_data = (test_x, test_y),
                    epochs = 20, 
                    batch_size = 5, 
                    callbacks = [EarlyStopping(patience=25),
                                 ModelCheckpoint('odds_loss.hdf5',
                                                 save_best_only=True)]
                   )
```

```python
print('Training Loss : {}\nValidation Loss : {}'.format(model.evaluate(train_x, train_y),
                                                        model.evaluate(test_x, test_y)))
```

```python
def accuracy_predictions(data, predict, house_bet_h, house_bet_a, house_bet_d):
    betted = 0
    accuracy = 0
    profits = 0
    for i in range(len(data)):
        if (data['FTR'][i] == 'H') and (data[predict][i] == 'H'):
            betted += 1
            accuracy += 1
            profits += (data[house_bet_h][i])

        if (data['FTR'][i] == 'A') and (data[predict][i] == 'A'):
            betted += 1
            accuracy += 1
            profits += (data[house_bet_a][i])

        if (data['FTR'][i] == 'D') and (data[predict][i] == 'D'):
            betted += 1
            accuracy += 1
            profits += (data[house_bet_d][i])
            print('Bet on draw! bet nº', betted+1)
            print(data['HomeTeam'][i], data['AwayTeam'][i], data['FTR'][i])
    
    print(' ')
    print('Accuracy = {}/{}'.format(accuracy, betted))
    print('Profits = {}'.format(profits))
    print('Win (+ or lost -) = ', profits-betted)
```
